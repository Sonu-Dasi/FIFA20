#### **Selecting Players with Overall score greater than 80**

```python
final_features = final_features[final_features.overall > 80]
names = final_features.short_name.tolist()
final_features.drop(['short_name'], axis = 1, inplace=True)
final_features.head()
```

#### **Scaling the data**

```python
from sklearn import preprocessing
```

```python
X = final_features.values # numpy array
scaler = preprocessing.MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_norm = pd.DataFrame(X_scaled)
X_norm.head()
```

#### **Reducing dimensions of the dataset**

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 1) # 2D PCA for the plot
reduced = pd.DataFrame(pca.fit_transform(X_norm))
reduced.head()
```

#### **Performing KMeans Clustering on the reduced dataset**

```python
from sklearn.cluster import KMeans

# specify the number of clusters
kmeans = KMeans(n_clusters=10)

# fit the input data
kmeans = kmeans.fit(reduced)

# predict the labels of clusters
labels = kmeans.predict(reduced)

#Getting unique labels
u_labels = np.unique(labels)
print(u_labels)

# centroid values
centroid = kmeans.cluster_centers_
print(centroid)

# cluster values
clusters = kmeans.labels_.tolist()
```

```python
reduced['cluster'] = clusters
reduced['name'] = names
#reduced['centroid'] = centroid
reduced.columns = ['X', 'cluster', 'name']
reduced.head() 
```

#### **Analysing the model performance and error**

```python
reduced.drop(['name'],axis=1,inplace=True)
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=50)
    kmeans.fit(reduced)
    wcss.append(kmeans.inertia_)
print(wcss)
plt.plot(range(1,11),wcss,marker='*')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
label=kmeans.labels_
label.shape
```

#### **Silhouette Score and Model Inertia**

```python
score=silhouette_score(reduced,label)
print('silhouette_score:',score)
```

```python
kmeans.inertia_
print('kmeans.inertia_:',kmeans.inertia_)
```
